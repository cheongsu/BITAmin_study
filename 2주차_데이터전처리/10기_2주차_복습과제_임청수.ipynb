### 2.레이블 인코딩
Seaborn 패키지에 있는 titanic 데이터를 이용해 레이블 인코딩을 진행해보도록 하겠습니다. <p>


```python
from sklearn.preprocessing import LabelEncoder

titanic = sns.load_dataset('titanic')
titanic
```

2-1) titanic에 있는 class 컬럼을 레이블 인코딩해서 c_labels에 저장해주세요



```python
# c_labels,index = pd.factorize(titanic['class'],sort=True)
# c_labels
encoder = LabelEncoder()
encoder.fit(titanic['class'])
c_labels = encoder.transform(titanic['class'])
c_labels
```

2-2) 레이블 인코딩해준 class 컬럼의 각 범주가 어떤 숫자로 변환되었는지 for문과 enumerate를 통해 확인해주세요.

```python
for i, j in enumerate(encoder.classes_):
  print(j,' -->',i)
```

2-3) 인코딩해준 값 c_labels을 다시 디코딩해주세요

```python
encoder.inverse_transform(c_labels)
```

### 4. Feature Scaling
마지막 Feature Scaling은 iris 데이터셋을 활용하여 표준화와 정규화를 진행하도록 하겠습니다.  <p>

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns= iris.feature_names)
iris_df
```

4-1) iris 데이터에 대해 표준화 하기 전 평균과 분산, 표준화하고 나서의 평균과 분산을 출력하시오.

```python

standard_scaler = StandardScaler()
#iris_df = iris_df.to_numpy()

print('표준화하기 전')
print('feature 평균')
print(iris_df.mean())
print('feature 분산')
print(iris_df.var())

print('표준화한 이후')
standard_scaler.fit(iris_df)
iris_scaled = standard_scaler.transform(iris_df)
scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print('feature 평균')
print(scaled.mean())
print('feature 분산')
print(scaled.var())
```

4-2) iris 데이터를 정규화를 한 뒤, 최대값과 최소값을 출력해주세요.

```python
scaler = MinMaxScaler()

scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)

print('feature 최소값')
print(iris_df_scaled.min())
print('feature 최대값')
print(iris_df_scaled.max())
```
