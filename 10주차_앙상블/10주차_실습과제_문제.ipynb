#### **2-3) Out of Bag (OOB) 점수**

OOB 점수를 출력하기 위해 기존 모델의 oob_score를 True로 지정해주세요. (기본값은 False)

```python
rf.set_params(warm_start=True,  '''문제''' )
```

**아래 코드는 실행하지 마시고 결과만 확인하세요. (오래 걸려요!!!)**

```python
#시간 체크용
start = time.time()

min_estimators = 200
max_estimators = 700

error_rate = {}

for i in range(min_estimators, max_estimators + 1):
    rf.set_params(n_estimators=i)
    rf.fit(x_train, y_train)

    oob_error = 1 - rf.oob_score_
    error_rate[i] = oob_error

end=time.time()
print('수행 시간: {0:.2f}'.format(end - start))

#그래프를 그리기 위해 dictionary를 pandas series로 변환 
oob_series = pd.Series(error_rate)

#plotting
import matplotlib.pyplot as plt
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10, 10))
oob_series.plot(kind='line', color = 'red')
plt.xlabel('n_estimators')
plt.ylabel('OOB Error Rate')
plt.title('OOB Error Rate')
```

- 트리 500개 즈음에서 OOB 오류율이 진동하기 시작하므로 모델의 n_estimators로 500을 사용하겠습니다.
- 이처럼 OOB 점수 시각화를 통해 n_estimators를 대략적으로 결정할 수 있습니다.
